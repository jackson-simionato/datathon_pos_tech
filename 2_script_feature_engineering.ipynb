## 2021

Não há nenhuma coluna com essa informação, apenas indica se é ingressante ou não. Opção é obter indiretamente dos outros dataframes

```python
df_2021.head()
```

```python
# Coluna com anos de PM do aluno em 2020 e 2022
df_passos_2021_anos_pm = df_2021.merge(df_2020[['NOME','ANOS_PM_2020']], on='NOME', how='left')
display(df_passos_2021_anos_pm.isna().sum())
df_passos_2021_anos_pm = df_passos_2021_anos_pm.merge(df_2022[['NOME','ANOS_PM_2022']], on="NOME", how='left')
display(df_passos_2021_anos_pm.isna().sum())
```

```python
import math

def calculate_anos_pm_2020(x):
    if not math.isnan(x):
        return int(x + 1)
    else:
        return None

def calculate_anos_pm_2022(x):
    if not math.isnan(x):
        return int(x - 1)
    else:
        return None
    
def calculate_anos_pm(x, y):
    if not math.isnan(x):
        return x
    else:
        if not math.isnan(y):
            return y
        else:
            return None
```

```python
# Anos PM a partir dos dados de 2020
df_passos_2021_anos_pm['ANOS_PM_2021_a'] = df_passos_2021_anos_pm['ANOS_PM_2020'].apply(calculate_anos_pm_2020)
```

```python
# Anos PM a partir dos dados de 2022
df_passos_2021_anos_pm['ANOS_PM_2021_b'] = df_passos_2021_anos_pm['ANOS_PM_2022'].apply(calculate_anos_pm_2022) 
```

```python
# Coluna final combinando duas colunas criadas
df_passos_2021_anos_pm['ANOS_PM_2021'] = df_passos_2021_anos_pm.apply(lambda row: calculate_anos_pm(row['ANOS_PM_2021_a'], row['ANOS_PM_2021_b']), axis=1)
```

```python
# Valor 0 para alunos novos
df_passos_2021_anos_pm.loc[df_passos_2021_anos_pm['SINALIZADOR_INGRESSANTE_2021'] == 'Ingressante', 'ANOS_PM_2021'] = 0
```

```python
# Apenas 4 alunos sem informação da contagem de anos
df_passos_2021_anos_pm['ANOS_PM_2021'].isna().sum()
```

```python
# Será utilizada abordagem de preenchimento com mediana
df_passos_2021_anos_pm['ANOS_PM_2021'].fillna(df_passos_2021_anos_pm['ANOS_PM_2021'].median(), inplace=True)
```

```python
# Adição da nova coluna no DF
list_values_anos_pm_2021 = df_passos_2021_anos_pm['ANOS_PM_2021']
df_2021.insert(2, 'ANOS_PM_2021', list_values_anos_pm_2021)
```

# Criação coluna INGRESSANTE_

```python
def get_ingressante(anos_pm):
    if anos_pm == 0:
        return 1
    else:
        return 0
    
df_2020['SINALIZADOR_INGRESSANTE_2020'] = df_2020['ANOS_PM_2020'].apply(get_ingressante)
df_2022['SINALIZADOR_INGRESSANTE_2022'] = df_2022['ANOS_PM_2022'].apply(get_ingressante)
```

# Criação coluna NIVEL_IDEAL_, DEFASAGEM

```python
def ajuste_fase_alfa(x):
    if x.startswith('ALFA'):
        return x.replace('ALFA ','FASE 0')
    else:
        return x
    
def calculo_nivel_ideal_int(nivel_ideal, nivel_atual):
    if nivel_ideal != 'Não se Aplica':
        nivel_str = nivel_ideal.split( )[1]
        return int(nivel_str)
    else:
        return int(nivel_atual)
```

# Merge DFs

```python
# Colunas em comum (2020-2021-2022): ['ID','NOME','ANO_PESQUISA','INSTITUICAO_ENSINO_ALUNO','BOLSISTA','ANOS_PM','SINALIZADOR_INGRESSANTE','FASE','TURMA','PONTO_VIRADA','INDE',
# 'PEDRA', 'IAA','IEG','IPS','IDA','IPP','IPV','IAN','NIVEL_IDEAL', 'DEFASAGEM']
```

```python
def generate_df_to_merge(df):
    common_cols = ['ID','NOME','ANO_PESQUISA','INSTITUICAO_ENSINO_ALUNO','BOLSISTA','ANOS_PM','SINALIZADOR_INGRESSANTE','FASE','TURMA','PONTO_VIRADA','INDE',
 'PEDRA', 'IAA','IEG','IPS','IDA','IPP','IPV','IAN','NIVEL_IDEAL', 'DEFASAGEM']
    
    ano = df['ANO_PESQUISA'].values[0]

    col_names = [col.split(f'_{ano}')[0] for col in df.columns]
    df.columns = col_names

    df_to_merge = df[common_cols]

    return df_to_merge
```

```python
df_2020_to_merge = generate_df_to_merge(df_2020)
df_2021_to_merge = generate_df_to_merge(df_2021)
df_2022_to_merge = generate_df_to_merge(df_2022)
```

```python
list_dfs = [df_2020_to_merge, df_2021_to_merge, df_2022_to_merge]

dfs_passos_massicos_merged = pd.DataFrame()

for df in list_dfs:
    if len(dfs_passos_massicos_merged) == 0:
        dfs_passos_massicos_merged = df.copy()
    else:
        dfs_passos_massicos_merged = pd.concat([dfs_passos_massicos_merged, df], axis=0)
        
dfs_passos_massicos_merged['NIVEL_IDEAL'] = dfs_passos_massicos_merged['NIVEL_IDEAL'].astype(float)
dfs_passos_massicos_merged['DEFASAGEM'] = dfs_passos_massicos_merged['DEFASAGEM'].astype(float)
dfs_passos_massicos_merged['ANOS_PM'] = dfs_passos_massicos_merged['ANOS_PM'].astype(int)
dfs_passos_massicos_merged.sort_values(by=['ANO_PESQUISA','ID'], inplace=True)
dfs_passos_massicos_merged.reset_index(drop=True, inplace=True)
```

```python
dfs_passos_massicos_merged.to_csv('./data/processed/passos_magicos/merged/df_passos_2020_2021_2022.csv', index=False)
```
